```python
from IPython.display import Javascript
import IPython



class HorseFrame(Javascript):
    
    def title(name, text = ""):
        return "<title> " + str(name) + " " + str(text) + " </title> "
    def head():
        return "<head> " + self.title + "</head> "
    
    def scripts(self,scripts,horse, list = None):
        start = "<script type='text/javascript'>\n" + horse.horseLinks(list) + "\n"
        for script in scripts:
            start = start + "\t\t" + script + "\n"
        
        end = "</script>"
        
        return start + end
         
    def __init__(self, horse , width=800, height=600, filter = None):
        
        self.width = width
        self.height = height
        self.horse = horse
        self.title = "<title> " + str(self.horse.name) + " </title> "
        l = []
        #print "list before", l
        #print filter is None, "fcheck"
        if filter is None:
            l = self.horse.races
        else:
            print "filter"+ str(filter)
            l = l + self.horse.filt(filter)
            
            
        #print "after", l
        self.scripts = self.scripts([fi,uif],horse, l)
        
        self.head = "<head> " + self.title + "\n" + self.scripts + "</head> "
        
        #print self.head
        
        
        
        #print type(self.title), type(self.head)
        
        
    

    def __repr__(self):
        
        for f in (self.head,horseCarousel,self.width, self.height,self.horse.href):
            print type(f)
        r = """
        <html>
               %s

            <body>
                %s
            
                <iframe
                    width="%i"
                    height="%i"
                    src="%s"
                    frameborder="0"
                    id = "hfr"
                ></iframe>
            </body>
        </html>
        
        """%(self.head,horseCarousel,self.width, self.height,self.horse.href)
        
        #print r
        
        return r
    
#start=int(timedelta(hours=1, minutes=46, seconds=40).total_seconds())

#filtered = hs[1].filt(byCourse)
hf = HorseFrame(hs[1],filter =byCourse)
#IPython.display.display(hf)

from IPython.display import display_html 


#help(hf)

```
